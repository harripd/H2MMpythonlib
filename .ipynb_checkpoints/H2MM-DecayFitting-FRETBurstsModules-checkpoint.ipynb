```python
print('Defining photon masks')
DAemDex_mask = d_all.get_ph_mask(ph_sel=Ph_sel(Dex='DAem'))
NanoTimes = d_all.nanotimes[0][DAemDex_mask]
AemDex_mask = d_all.get_ph_mask(ph_sel=Ph_sel(Dex='Aem'))
AemDex_mask_red = AemDex_mask[DAemDex_mask]

d_all_bursts = d_all.mburst[0]
ph_d = d_all.get_ph_times(ph_sel=Ph_sel(Dex='DAem'))
print('Reducing photon burst indexes')
burst_red = d_all_bursts.recompute_index_reduce(ph_d)

TotalArrivalTime = np.empty([1,0], dtype=int)
ArrivalTime = []
ArrivalNanoTime = []
ArrivalColor = []

nphot = 0
print('Separating data into bursts')
for start, stop in zip(burst_red.istart, burst_red.istop+1):
    ArrivalTime.append(ph_d[start:stop])
    ArrivalNanoTime.append(NanoTimes[start:stop])
    ArrivalMask_temp = AemDex_mask_red[start:stop]
    ArrivalColor_temp = np.zeros(ArrivalMask_temp.size,dtype=int)
    ArrivalColor_temp[ArrivalMask_temp] = 1
    ArrivalColor.append(ArrivalColor_temp)
    nphot += ArrivalColor_temp.size
print(f'Number of bursts: {burst_red.num_bursts}, Number of Photons: {nphot}')
```

Now the 2 state model is defined and optimized

```python
vb_fit = viterbi_all(h_mod2,ArrivalColor,ArrivalTime)
path = vb_fit[0]
prob = vb_fit[1]
```

```python
dd_low_mask = [np.zeros(ArrivalColor[b].shape,dtype=bool) for b in range(0,nburst)]
dd_low_nanotime = np.zeros(0, dtype=int)
da_low_mask = [np.zeros(ArrivalColor[b].shape,dtype=bool) for b in range(0,nburst)]
da_low_nanotime = np.zeros(0, dtype=int)
dd_high_mask = [np.zeros(ArrivalColor[b].shape,dtype=bool) for b in range(0,nburst)]
dd_high_nanotime = np.zeros(0, dtype=int)
da_high_mask = [np.zeros(ArrivalColor[b].shape,dtype=bool) for b in range(0,nburst)]
da_high_nanotime = np.zeros(0, dtype=int)

min_prob = 0.9
for b in range(0,nburst):
    dd_low_mask[b] = (path[b] == 1) & (prob[b] > min_prob) & (ArrivalColor[b] == 0)
    dd_low_nanotime = np.append(dd_low_nanotime, ArrivalNanoTime[b][dd_low_mask[b]])
    da_low_mask[b] = (path[b] == 1) & (prob[b] > min_prob) & (ArrivalColor[b] == 1)
    da_low_nanotime = np.append(da_low_nanotime, ArrivalNanoTime[b][da_low_mask[b]])
    dd_high_mask[b] = (path[b] == 0) & (prob[b] > min_prob) & (ArrivalColor[b] == 0)
    dd_high_nanotime = np.append(dd_high_nanotime, ArrivalNanoTime[b][dd_high_mask[b]])
    da_high_mask[b] = (path[b] == 0) & (prob[b] > min_prob) & (ArrivalColor[b] == 1)
    da_high_nanotime = np.append(da_high_nanotime, ArrivalNanoTime[b][da_high_mask[b]])
```

```python
tcspc_bin = d_all.nanotimes_params[0]['tcspc_unit']
tcspc_range = d_all.nanotimes_params[0]['tcspc_range']
tcspc_num_bins = d_all.nanotimes_params[0]['tcspc_num_bins']
bins_raw = np.arange(0, tcspc_num_bins)

hist_params = dict(bins=bins_raw, density=False)
dd_low_hist = np.histogram(dd_low_nanotime,**hist_params)
da_low_hist = np.histogram(da_low_nanotime,**hist_params)
dd_high_hist = np.histogram(dd_high_nanotime,**hist_params)
da_high_hist = np.histogram(da_high_nanotime,**hist_params)
```

```python
time_axis_ns = (bins_raw[1:] * tcspc_bin) * 1e9

plt.semilogy(time_axis_ns, dd_low_hist[0], 'k')
plt.semilogy(time_axis_ns, da_low_hist[0], 'b')
plt.semilogy(time_axis_ns, dd_high_hist[0], 'g')
plt.semilogy(time_axis_ns, da_high_hist[0], 'r')
plt.xlim(27,49)
plt.xlabel('Time (ns)')
plt.ylabel('Counts')
```

Calculating BIC' as described in Lerner et. al. JPC 2018, 148, 123315

```python
bic = np.zeros((4),dtype=float)
Kstate = np.zeros((4),dtype=int)
Kstate = [q**2 + q - 1 for q in range(1,5)]
lnphot = np.log(nphot)
bic[0] = -2*h_mod1.loglik + Kstate[0]*lnphot
bic[1] = -2*h_mod2.loglik + Kstate[1]*lnphot
bic[2] = -2*h_mod3.loglik + Kstate[2]*lnphot
bic[3] = -2*h_mod4.loglik + Kstate[3]*lnphot
dbic = bic - bic.min()
bicp = [dbic[i]/(nphot-Kstate[i]) for i in range(0,4)]
bicp
```
